```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("/content/winequalityProject.csv")
data
```

```python
#Rename columns name to simplification
data.rename(columns={'fixed acidity':'Fix_Acidity'},inplace=True)
data.rename(columns={'volatile acidity':'Volatile_Acid'},inplace=True)
data.rename(columns={'citric acid':'Citric_Acid'},inplace=True)
data.rename(columns={'residual sugar':'Residual_Sugar'},inplace=True)
data.rename(columns={'free sulfur dioxide':'Free_s_oxide'},inplace=True)
data.rename(columns={'total sulfur dioxide':'Total_s_oxide'},inplace=True)
data
```

```python
#Coversion of data
data.Fix_Acidity = data.Fix_Acidity.astype(float)
data.Volatile_Acid = data.Volatile_Acid.astype(float)
data.Citric_Acid = data.Citric_Acid.astype(float)
data.Residual_Sugar = data.Residual_Sugar.astype(float)
data.Free_s_oxide = data.Free_s_oxide.astype(float)
data.Total_s_oxide = data.Total_s_oxide.astype(float)
data
```

```python
columns_to_clean = ['Fix_Acidity', 'Volatile_Acid', 'Citric_Acid','Residual_Sugar','Free_s_oxide','Total_s_oxide','chlorides','density','pH','sulphates','alcohol','quality']
data[columns_to_clean] = data[columns_to_clean].fillna(0)
```
